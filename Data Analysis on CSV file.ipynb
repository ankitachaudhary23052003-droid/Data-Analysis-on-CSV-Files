```python
import pandas as pd
import matplotlib.pyplot as plt

# Show plots inside notebook
%matplotlib inline
```

```python
import os
os.listdir()
df = pd.read_csv("retail_sales_dataset.csv")
df.head()

```

```python
data = {
    "Date": ["2026-01-01","2026-01-02","2026-01-03","2026-01-04","2026-01-05",
             "2026-01-06","2026-01-07","2026-01-08","2026-01-09","2026-01-10"],
    "Region": ["North","South","East","West","North","South","East","West","North","South"],
    "Product": ["Laptop","Mobile","Tablet","Laptop","Mobile","Tablet","Laptop","Mobile","Tablet","Laptop"],
    "Sales": [50000,30000,20000,45000,35000,25000,60000,40000,22000,48000]
}

df = pd.DataFrame(data)
df.to_csv("sales_data.csv", index=False)

df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
total_sales = df["Sales"].sum()
print("Total Sales:", total_sales)
```

```python
region_sales = df.groupby("Region")["Sales"].sum()
print(region_sales)
```

```python
region_sales.plot(kind="bar")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()
```

```python
product_sales = df.groupby("Product")["Sales"].sum()
print(product_sales)
```

```python
product_sales.plot(kind="bar")
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.show()
```

```python
highest_region = region_sales.idxmax()
print("Highest Selling Region:", highest_region)
```

```python
highest_product = product_sales.idxmax()
print("Highest Selling Product:", highest_product)
```
